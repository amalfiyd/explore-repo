```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
train_file = r"C:\Users\amalf\OneDrive\Projects\05 - HandsonML\Dataset\titanic\train_data.csv"
test_file = r"C:\Users\amalf\OneDrive\Projects\05 - HandsonML\Dataset\titanic\test_data.csv"
```

```python
df_train = pd.read_csv(train_file)
df_test = pd.read_csv(test_file)
```

```python
df_train.head()
```

```python
df_train.drop(['Unnamed: 0', 'PassengerId'], axis=1, inplace=True)
df_test.drop(['Unnamed: 0', 'PassengerId'], axis=1, inplace=True)
```

```python
df_train.describe()
```

```python
def dataframe_pipeline(df):
    df_temp = df.copy()
    cols_1 = ['Pclass_1','Pclass_2','Pclass_3']
    cols_2 = ['Title_1','Title_2','Title_3','Title_4']
    cols_3 = ['Emb_1','Emb_2','Emb_3']
    df_temp['Pclass'] = df_temp.apply(lambda x: np.argmax(x[cols_1].values) + 1, axis=1)
    df_temp['Title'] = df_temp.apply(lambda x: np.argmax(x[cols_2].values) + 1, axis=1)
    df_temp['Emb'] = df_temp.apply(lambda x: np.argmax(x[cols_3].values) + 1, axis=1)
    
    return df_temp.drop(cols_1 + cols_2 + cols_3, axis=1)
```

```python
df_train_prep = dataframe_pipeline(df_train)
df_test_prep = dataframe_pipeline(df_test)

y_train = df_train_prep.Survived
y_test = df_test_prep.Survived

df_train_prep.drop(['Survived'], axis=1, inplace=True)
df_test_prep.drop(['Survived'], axis=1, inplace=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators = [100, 300, 500, 800, 1000]
criterion = ['gini', 'entropy']
bootstrap = [True, False]

pipeline = Pipeline([
    ('std_scaler', RobustScaler()),
    ('predictor', RandomForestClassifier()),
)

gs_param= {
    'predictor__random_state': 42,
    'predictor__n_estimators': n_estimators,
    'predictor__criterion': criterion,
    'predictor__bootstrap': bootstrap,
}

rfc = RandomForestClassifier(random_state=42)
```

```python
CV = GridSearchCV(estimator=rfc, param_grid=gs_param, cv=5)
```

```python
CV.fit(df_train_prep, y_train)
```

```python
print(CV.best_score_)    
print(CV.best_params_)    
```
